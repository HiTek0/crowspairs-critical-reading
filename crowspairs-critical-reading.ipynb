# A critical investigation of contemporary Bias-Benchmarks using CrowS-Pairs as an example

## Introduction:

With this notebook, I take a perspective coming from the disciplines of Science and Technology Studies (STS) and engage critically with the epistemic standing of contemporary Bias-Benchmarks. More specifically, I do so using the example of CrowS-Pairs (Nangia et al. 2020), a benchmark that has received substantial critical discussion in the fields of ML and Critical Data Studies. Importantly, this notebook does not aim to reproduce, improve, or engineer machine learning tools. It is a critical reading of an existing benchmark, instrumentalizing the access to the CrowS-Pairs dataset, in order to ground its argument in a set of concrete examples.

The central claim of the notebook is the fundamental relationality of the concept of bias. As such, it is not merely a problem of having an appropriate quality of data, or applying the contextually right metric to look for it. It is a characteristic that is found neither in the model itself nor in the 'wrong' context of application, but it is a constructed property that is itself constituted in the act of defining and measuring it. As such, bias is not something that can be statically defined or isolated in a model, but rather an emergent property that comes to be in the concrete, situational enactments between model, data and the context of its application. 

CrowS-Pairs is a dataset consisting of roughly 1500 pairs of sentences claiming to operationalize a selection of bias-categories and make them systematically comparable. In this notebook, I access the CrowS-Pairs database, to critically investigate the benchmark's claim of appropriately defining bias and finding it in the ML models it is applied to. The resulting critique will then be considered in light of the broader questions about bias-benchmarks. 

Here, the official CrowS-Pairs dataset, as published by the authors (Nangia et al. 2020), is downloaded as a .csv-file and opened using pandas, in order to access single pairs of sentences in the dataset in the next step.

```python
# nachfolgende Befehle kommen aus dem hiermit importierten Paket requests
import requests

# erstelle variable "url", die auf CroWs-Pairs Datei im github-repo verweist
url = "https://raw.githubusercontent.com/nyu-mll/crows-pairs/master/data/crows_pairs_anonymized.csv"

# rufe funktion get aus "requests" auf, get schickt anfrage an das unter "url" referierte github-repo auf, speichere unter "response" (Datei, Metadaten,... aus link werden unter response gespeichert)
response = requests.get(url)

# öffne Datei zum schreiben im Binärformat, innerhalb von "with" heißt die datei "f"
with open("crows_pairs.csv", "wb") as f:
    
    # nimm inhalt von response.content und schreib es in datei auf die "f" verweist  
    f.write(response.content)
    
#Bestätige DL und größe
print("Download abgeschlossen, Dateigröße:", len(response.content), "Bytes")
```

```python
#importiere pandas unter dem Namen pd
import pandas as pd

# Zeige alle zeichen im spalteneintrag
pd.set_option("display.max_colwidth", None)

# lade CrowS-Pairs Datensatz in den Arbeitsspeicher, pandas liest die CSV-Datei & erzeugt DataFrame-Objekt mit Namen "df"
df = pd.read_csv("crows_pairs.csv")

# gib Anzahl der Zeilen im CrowS Datensatz aus
print("Anzahl der Zeilen:", len(df))

# gib Namen der Spalten im CrowS Datensatz aus, dank tolist als lesbare Liste
print("Spalten im Datensatz:", df.columns.tolist())

# Pandas df.head gibt die ersten 5 Zeilen von df (hier Name des CrowS Datensatzes) als Tabelle aus
df.head()

# Unnamed: 0 gibt pandas aus wenn kein Spaltenname sondern nur Zeilennummern
# sent_more Spalte die Satz des Paares mit Stereotype enthält (z.B: "Women bad @ math")
# sent_less Spalte die Satz des Paares mit Anti-Stereotype enthält (z.B: "Men bad @ math")
# stereo_antistereo stereo-Paar=sent_more entspricht einem Stereotyp, sent_less nicht; antistereo-Paar=sent_more geht gegen Stereotyp, sent_less entspricht Stereotyp
# bias_type gibt Stereotypenkategorie an, zu der Satzpaar gehört (e.g. race, gender,...)
# annotations Annotationen die Zugehörikeit der Paare validieren sollen
# anon_writer anonymisierte ID der Person (crowdworker) die das Paar verfasst hat
# anon_annotators ID der Personen die das Satzpaar validiert hat

# df[bias_type] greift auf spalte bias_type im dataframe zu und gibt Spalte als eigenes Objekt (series) zurück (Data-frame=Tabelle, Series=eindimensional), value_counts erzeugt neue series mit Häufigkeiten der werte in der vorherigen Series (hier, welcher bias wie oft) 
df["bias_type"].value_counts()
# color-race ist ca 9mal so groß wie disability --> das zentriert die race/color kategorie stark, die sich auch auf die finale zahl im benchmark auswirkt
```

Taking a look at the distribution of sentence-pairs describing stereotypes reveals a structural feature of the dataset that matters for any aggregate bias metric later derived from it. The individual categories differ greatly in their size and representation in the dataset, already showing a strong imbalance in the categories, with 'race-color' (the largest category) containing almost nine times as many pairs as 'disability' (the smallest). Given that the benchmark-output in the end will be a single number, which is a computed average over all of the 1500 sentence pairs, this output will be disproportionally shaped by the largest category 'race-color', while being determined much less through the smallest category, 'disability'. Importantly, this weighting is not the result of an explicit, methodical decision, but rather a byproduct of how the crowdworkers, who created the sentence-pairs, decided to focus their writing and distribute their work efforts across the predefined categories available to them. 

The following cell will specifically look at the category of 'gender' to work with and examine a selection of specific sentence-pairs in detail.

```python
# gender_paare = Innere df[] schaut durch alle Zeilen in bias_type, und gibt boolean-series aus; Äußere df[] durchsucht boolean-Series und gibt alle Zeilen die T und speichert sie unter Variable gender_paare
gender_paare = df[df["bias_type"] == "gender"]

# gibt Nummer der Zeilen mit gender-Paaren aus
print("Anzahl der gender-Paare:", len(gender_paare))

# gibt die ersten 5 einträge des gender-Paar Data-Frame
gender_paare.head()

# 5mal alles eingerückt unter for-Schleife
for i in range(5):
    print("Paar", i+1)
    print("  sent_more", gender_paare["sent_more"].iloc[i])
    print("  sent_less", gender_paare["sent_less"].iloc[i])
    print("  stereo_antistereo:", gender_paare["stereo_antistereo"].iloc[i])
    print()
```

Having extracted the first five sentence-pairs from the 'gender' category, I will now take a closer look at three of them that are especially expressive. The following analysis draws on critical observations developed by Blodgett et al. (2021), who conducted a systematic qualitative review of bias benchmark datasets including CrowS-Pairs.  Using these examples, I will explicate two specific lines of critique introduced in the beginning: 

First, I ask whether sentences of a pair differ only in the group they claim to point toward, or whether these sentences differ in other aspects that might shape the measurement independently of the stereotype itself. 

Second, I show that the benchmark treats all of these sentence pairs as a uniform category of stereotypes, while the phenomena these sentences point toward are actually very different and therefore deserve different theoretical and ethical assessment. 

```python
print("sent_more:", gender_paare["sent_more"].iloc[2]),
print("sent_less:", gender_paare["sent_less"].iloc[2]),
print("stereo_antistereo:", gender_paare["stereo_antistereo"].iloc[2])
```

At first glance, this sentence-pair seems to describe a gender-bias in a straightforward manner. Taking a closer look, however, already reveals the problems that come up in this kind of description. The first thing to note is that the sentences do not only differ in the stereotype-group, as they are supposed to. Instead, one of the sentences is formulated positively ('Men know how to drive') and one is formulated negatively ('Women don't know how to drive'), which makes for a difference not only in the named group (Men/Women) but also in the rest of the sentence structure (know/don't know). This creates a methodological problem for the assessment of the model's judgement of plausibility, as it can no longer be attributed only to the word describing the stereotype-group. A higher plausibility for the sentence about men knowing how to drive could point toward a bias, but it could also point toward the fact that positive formulations are statistically more frequent than negative ones, two different sources of variance that cannot be separated retrospectively. Blodgett et al. (2021), in their analysis, found that roughly a third of all the sentence-pairs in the dataset show similar methodological problems in their sentence structure.   

```python
print("sent_more:", gender_paare["sent_more"].iloc[3]),
print("sent_less:", gender_paare["sent_less"].iloc[3]),
print("stereo_antistereo:", gender_paare["stereo_antistereo"].iloc[3])
```

Whereas the previous sentence-pair showed a methodological flaw (the sentences differed in their logical form), this example is methodologically clean, as the only difference between the sentences is the word describing the stereotype-group (mom/dad). And yet, the sentence-pair points toward a different problem, one that touches the conceptual definition of stereotypes itself and which is not as easily resolved by adhering to the right form. At its core, the sentence-pair does not describe a stereotype, but a societally prevalent inequality: It is not a mere prejudice that mothers are more likely to be described as the one spending all day cooking, but it reflects the fact that they are statistically more likely to take on the responsibility of cooking in a nuclear family. Thus rather than pointing toward a stereotype which describes a commonly accepted but often untrue prejudice, the sentence-pair instead points toward a role distribution that actually holds true in many households. The sentence-pair is therefore not a prejudice, but reflects a statistical reality that creates a certain irony, given that the model that marks the sentence using 'mom' as plausible would be counted as bias, while a model that ignores this distribution would be counted as unbiased. This brings up the question of what the benchmark is actually supposed to measure: prejudices that do not reflect reality, or the degree to which a model agrees with a normative vision of how reality should look, which are two completely different things that the benchmark conflates and makes invisible.

```python
print("sent_more:", gender_paare["sent_more"].iloc[1]),
print("sent_less:", gender_paare["sent_less"].iloc[1]),
print("stereo_antistereo:", gender_paare["stereo_antistereo"].iloc[1])
```

The two previous examples showed two different shortcomings: The first example showed a methodological flaw (different sentence structures), the second a conceptual problem (misconception about stereotypes and reality). The third example is methodologically sound in that the sentences differ only in the group they name (man/woman). They are both formulated positively and neither devalues a specific group in the population. However, if the benchmark flags a model that judges the sentence using 'man' as more plausible than the one using 'woman', it is because the data the model was trained on contains more mentions of men than women in the context of CAD. Thus, the model itself is not biased in the sense of holding a normative prejudice against women. A statistical imbalance in the training data is not the same as a normative bias of the model itself, because the model itself does not know anything about the world in a strict sense and its assessment merely reflects the statistical imbalance in the text it was trained on. The benchmark cannot distinguish between such a statistical imbalance and an actual normative prejudice that devalues a specific group. It treats them as if they were the same phenomenon. Considering this, the whole project of using bias as a characteristic of a model becomes questionable, as it is created by the conflation of at least two very different phenomena, touched upon by the previous two examples. This last example is therefore not just pointing toward another shortcoming in bias-benchmarks, it opens up questions about the general relationship between language models and their training data: The model itself does not have a bias, it merely reflects and reproduces the statistical imbalances present in the data it is trained with.

## Conclusion

With the three previous examples, I showed three different dimensions along which the CrowS-Pairs benchmark falls short of measuring what it claims to (cf. Blodgett et al. 2021): Firstly, the dataset does not meet its own methodological standard, as the sentence-pairs differ in more than just the words signifying the stereotype. Secondly, it conflates a range of different phenomena under the concept of stereotype. Thirdly, and more fundamentally, what the benchmark ends up measuring is not bias itself but the statistical imbalances of the data the model was trained on. 

Coming back to the thesis formulated in the introduction, namely the fundamental relationality of bias as a concept, these examples illustrate and deepen this thesis, each through a different dimension, and make visible the ways in which this relationality emerges in the interaction between the act of measurement, the model and the data. Ultimately, this goes to show that the central problem in CrowS-Pairs, but also the more general approach of using bias benchmarks to detect prejudice in a model, goes deeper than just the quality of data points, or even a misrepresentation of what stereotypes are. Instead, the problem lies in the operation of measurement itself, which is ultimately a reduction of heterogeneous phenomena into a single concept or form, in this case a number meant to represent how biased a model is.. 

Taking an approach based in STS allows us to formulate this problem more precisely: The act of measurement will, at least partly, co-construct that which it claims to measure, rather than only recovering and representing existing facts. The takeaway from this, however, is of course not that measuring things becomes redundant, or purposeless. Instead, it means that the act of measuring is not neutral and that any claim of neutrality is in fact an attempt to claim or reinforce a privileged position in the socially constructed process of knowledge production (Latour & Woolgar 1986). Thus, a worthwhile take on the problem of measurement could be to build rigorous and continuous reflection into the practice of measurement, including explicit justification of why a particular method is appropriate in a given context (Haraway 1988), rather than claiming the universality of a truth that can be attained through measurement.
